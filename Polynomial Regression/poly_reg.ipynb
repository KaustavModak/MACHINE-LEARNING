```python
# importing libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
X=6 * np.random.rand(100,1)-3
y=0.5 * X**2 + 1.5*X +2+np.random.rand(100,1)
plt.scatter(X,y,color='r')
plt.xlabel('X dataset')
plt.ylabel('Ydataset')
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.25,random_state=42)
```

```python
# implementing simple linear regression
from sklearn.linear_model import LinearRegression
regression1=LinearRegression()
```

```python
regression1.fit(X_train,y_train)
```

```python
from sklearn.metrics import r2_score
score1=r2_score(y_test,regression1.predict(X_test))
print(score1)
```

```python
# visualizing this model
plt.scatter(X_train,y_train)
plt.plot(X_train,regression1.predict(X_train),color='r')
plt.xlabel('X dataset')
plt.ylabel('Y')
```

```python
# lets apply polynomial transformation
from sklearn.preprocessing import PolynomialFeatures
```

```python
poly=PolynomialFeatures(degree=2,include_bias=True)
# without include_bias : h(x)=b0 + b1(x1) + b2(x1)^2
# with include_bias : h(x)=b0*1 +b1(x1) + b2(x1)^2
X_train_poly=poly.fit_transform(X_train)
X_test_poly=poly.fit_transform(X_test)
```

```python
X_train_poly
```

```python
X_test_poly
```

```python
regression=LinearRegression()
regression.fit(X_train_poly,y_train)
y_pred=regression.predict(X_test_poly)
score=r2_score(y_test,y_pred)
print(score)
```

```python
print(regression.coef_)
```

```python
print(regression.intercept_)
```

```python
plt.scatter(X_train,regression.predict(X_train_poly))
plt.scatter(X_train,y_train)
```

```python
# prediction of new data
X_new=np.linspace(-3,3,200).reshape(200,1)
X_new_poly=poly.transform(X_new)
X_new_poly
```

```python
y_new=regression.predict(X_new_poly)
plt.plot(X_new,y_new,"r-",linewidth=2,label="New Predictions")
plt.plot(X_train,y_train,"b.",label="Training points")
plt.plot(X_test,y_test,"g.",label="Testing points")
plt.xlabel("X")
plt.ylabel("Y")
plt.legend()
plt.show()
```
